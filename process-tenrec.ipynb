```python
import pandas as pd 
import numpy as np
import torch
from sklearn.preprocessing import LabelEncoder
inter_data_original = pd.read_csv('./QB-video.csv', header = 0)
```

```python
print(inter_data_original.describe())
```

```python
data = pd.DataFrame({})
data['user_id'] = inter_data_original['user_id']
data['item_id'] = inter_data_original['item_id']
data['ratings'] = inter_data_original['click']
data['gender'] = inter_data_original['gender']
print(data.head(5))
print(data.shape)
print(pd.isnull(data).any())
```

```python
print(data['user_id'].value_counts())
cnt = data['user_id'].value_counts()
users_50 = list(cnt[cnt>=50].index.tolist())
```

```python
data = data.loc[data['user_id'].isin(users_50), :]
cnt = data['item_id'].value_counts()
items_10 = list(cnt[cnt>=10].index.tolist())
data = data.loc[data['item_id'].isin(items_10), :]
print(data.head(5))
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
data['user_id'] = lbe.fit_transform(data['user_id'])
data['item_id'] = lbe.fit_transform(data['item_id'])
print(data.describe())
```

```python
data = data[data['gender']!=2]
```

```python
print(data)
print(data['gender'].describe()) 
```

```python
print(data.groupby('gender').describe())
```

```python
data['ratings'] = data['ratings'].apply(lambda x: 5 if x == 1 else 0)
```

```python
print(data.describe())
```

```python
data.reset_index(inplace=True, drop=True)
data.to_hdf('./tenrec.h5', key = '50core')
```
